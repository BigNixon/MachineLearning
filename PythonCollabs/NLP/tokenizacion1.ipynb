```python
import re
```

```python
exp = re.compile(r"\d\d\d").findall("varios numeros: 988, 222, 333, 987654321, 2, 12")
exp
```

```python
exp = re.compile(r"\s\d\d\d").findall("varios numeros: 988, 222, 333, 987654321, 2, 12")
exp
```

```python
exp = re.compile(r"\s\d+").findall("varios numeros: 988, 222, 333, 987654321, 2, 12")
exp
```

```python
exp = re.compile(r"\s\d+|,\d+").findall("varios numeros: 988, 222, 333, 987654321, 2, 12")
exp
```

```python
exp = re.compile(r":\)|:-\)").findall("estoy feliz :) :-)") #escapamos el caracter especia ")" con "\"
exp
```

```python
#DETECTAR PALABRAS QUE EMPIEZAN CON "C"
exp1 = re.compile(r"\Ac\w+|\sc\w+").findall("como estaba al principio y contaba al compás") 
exp2 = re.compile(r"\Ac[a-zA-ZáéíóúÁÉÍÓÚ]+|\sc[a-zA-ZáéíóúÁÉÍÓÚ]+").findall("como estaba al principio y contaba al compás") 
print(exp1, exp2)
```

```python
#DETECTAR UN NUMERO CON DECIMALES #.#
exp = re.compile(r"\d+\.\d+") # tambien puede ser "\d+[.]\d+"
exp.findall("te devuelvo los 5630.25 soles que me prestaste")
```

```python
#DETECTAR VALORES NUMERICOS QUE PUEDEN O NO TENER VALORES DECIMALES
exp = re.compile(r"\d+\.\d+ | \d+") 
exp.findall("te devuelvo los 5630.25 y los 120 soles que me prestaste")
```

```python
#DETECTAR VALORES NUMERICOS CON SEPARADORES DE MILES O MILLARES
exp = re.compile(r"(\d+(,\d\d\d)*([.]\d+)?)") 
exp.findall("te devuelvo los 5,630,426.25 y los 120,25 soles que me prestaste")
```
